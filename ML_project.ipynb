## Data Visualization

```python
import random
connections = [
    (0, 1), (1, 2), (2, 3), (3, 4),       # Thumb
    (0, 5), (5, 6), (6, 7), (7, 8),       # Index
    (0, 9), (9, 10), (10, 11), (11, 12),  # Middle
    (0, 13), (13, 14), (14, 15), (15, 16),# Ring
    (0, 17), (17, 18), (18, 19), (19, 20),# Pinky
    (5, 9), (9, 13), (13, 17)             # Palm/Knuckles
]
```

### Visualizing samples before normalizing

```python
def sample_visualize(df, no_of_samples):
  for sample in range(no_of_samples):
    i = random.choice(df.index)
    row1 = df.iloc[i,:-1]
    row1_label = df.iloc[i,-1]
    row1_vis = row1.values.reshape(21,3)
    X = row1_vis[:,0]
    Y = row1_vis[:, 1]
    plt.figure(figsize=(8,6))
    plt.scatter(x = X, y = Y, label = row1_label)
    plt.gca().invert_yaxis()
    for connection in connections:
      plt.plot([X[connection[0]], X[connection[1]]], [Y[connection[0]], Y[connection[1]]], 'rs-')
    plt.legend()
    plt.show()
sample_visualize(df,5)
```

# Data Preprocessing

### Data normalization: Recentering
- recentering wrist to be the origin so that hand cooridnates become relative to wrist position instead of being relative to image frame origin
- this helps the model to learn the shape of gesture itself and rule out the factor of geusture position -> so a 'peace' sign in the top right position is considered the same as 'peace' sign in the bottom left position

```python
X_columns = df.columns[['x' in col for col in (df.columns)] ]
```

```python
df[X_columns] = df[X_columns].sub(df['x1'], axis = 0)
```

```python
Y_columns = df.columns[['y' in col for col in df.columns]]
Y_columns
```

```python
df[Y_columns] = df[Y_columns].sub(df['y1'], axis = 0)
```

```python
df
```

### Normalization:
- normalizing all landmarks by dividing by hand length(equlidian distance between middle finger tip and wrist)
- this rules out the difference in cooridantes that happens due to different hand size or how far geusture is from the camera
- this also helps the Model to focus on the shape of geusture only and learn it , without making irrelevant features like proximity or size affect the learning process  

```python
df['dist'] = np.sqrt(np.square(df['x12']) + np.square(df['y12']))
df['dist']
```

```python
df['dist'] = np.sqrt(np.square(df['x12']) + np.square(df['y12']))
df['dist'].replace(0,1) # -> to prevent dividing by zero
df[list(X_columns)+list(Y_columns)] = df[list(X_columns)+list(Y_columns)].div(df['dist'], axis=0)
```

```python
df.drop(['dist'], axis = 1, inplace = True)
```

```python
df
```
