```python
import random
connections = [
    (0, 1), (1, 2), (2, 3), (3, 4),       # Thumb
    (0, 5), (5, 6), (6, 7), (7, 8),       # Index
    (0, 9), (9, 10), (10, 11), (11, 12),  # Middle
    (0, 13), (13, 14), (14, 15), (15, 16),# Ring
    (0, 17), (17, 18), (18, 19), (19, 20),# Pinky
    (5, 9), (9, 13), (13, 17)             # Palm/Knuckles
]
```

### Visualizing samples before normalizing

```python
def sample_visualize(df, no_of_samples):
  for sample in range(no_of_samples):
    i = random.choice(df.index)
    row1 = df.iloc[i,:-1]
    row1_label = df.iloc[i,-1]
    row1_vis = row1.values.reshape(21,3)
    X = row1_vis[:,0]
    Y = row1_vis[:, 1]
    plt.figure(figsize=(8,6))
    plt.scatter(x = X, y = Y, label = row1_label)
    plt.gca().invert_yaxis()
    for connection in connections:
      plt.plot([X[connection[0]], X[connection[1]]], [Y[connection[0]], Y[connection[1]]], 'rs-')
    plt.legend()
    plt.show()
sample_visualize(df,5)
```

# Data Preprocessing

### Data normalization: Recentering
- recentering wrist to be the origin so that hand cooridnates become relative to wrist position instead of being relative to image frame origin
- this helps the model to learn the shape of gesture itself and rule out the factor of geusture position -> so a 'peace' sign in the top right position is considered the same as 'peace' sign in the bottom left position

```python
X_columns = df.columns[['x' in col for col in (df.columns)] ]
```

```python
df[X_columns] = df[X_columns].sub(df['x1'], axis = 0)
```

```python
Y_columns = df.columns[['y' in col for col in df.columns]]
Y_columns
```

```python
df[Y_columns] = df[Y_columns].sub(df['y1'], axis = 0)
```

```python
df
```

### Normalization:
- normalizing all landmarks by dividing by hand length(equlidian distance between middle finger tip and wrist)
- this rules out the difference in cooridantes that happens due to different hand size or how far geusture is from the camera
- this also helps the Model to focus on the shape of geusture only and learn it , without making irrelevant features like proximity or size affect the learning process  

```python
df['dist'] = np.sqrt(np.square(df['x12']) + np.square(df['y12']))
df['dist']
```

```python
df['dist'] = np.sqrt(np.square(df['x12']) + np.square(df['y12']))
df['dist'].replace(0,1) # -> to prevent dividing by zero
df[list(X_columns)+list(Y_columns)] = df[list(X_columns)+list(Y_columns)].div(df['dist'], axis=0)
```

```python
df.drop(['dist'], axis = 1, inplace = True)
```

```python
df
```

### Visalization after Normalizing: scale changed, also samples of same class becomes in similar cooridantes

# Data Splitting

```python
from sklearn.model_selection import train_test_split
X = df.drop(['label'], axis = 1)
y = df['label']
```

```python
y.value_counts()
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,y, stratify=y, test_size=0.2)
```

```python
X_val,X_test, y_val, y_test = train_test_split(X_test, y_test, stratify=y_test, test_size=0.5)
```

# Model training and evaluation

```python
from sklearn import metrics
def eval_metrics(actual, pred):
    accuracy = metrics.accuracy_score(actual, pred)
    f1 = metrics.f1_score(actual, pred, average='weighted')
    prec = metrics.precision_score(actual, pred, average='weighted')
    recall = metrics.recall_score(actual, pred, average='weighted')
    return accuracy, f1, prec, recall
```

```python
def eval_model(model, y_true, X):
  y_true = y_true
  y_pred = model.predict(X)
  acc, f1, prec, recall = eval_metrics(y_true, y_pred)
  print(f'Model Accuracy: {acc:.4f},\n F1-Score: {f1:.4f},\n Precision: {prec:.4f},\n Recall: {recall:.4f}')
```

## XGBOOST RUNS

```python
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBClassifier

le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)
y_val_enc = le.transform(y_val)
y_test_enc = le.transform(y_test)
```

```python
def eval_model_numeric_classes(model, label_encoder, y_true, X):
  le = label_encoder
  y_true = y_true
  y_pred_enc = model.predict(X)
  y_pred = le.inverse_transform(y_pred_enc)
  acc, f1, prec, recall = eval_metrics(y_true, y_pred)
  print(f'Model Accuracy: {acc:.4f},\n F1-Score: {f1:.4f},\n Precision: {prec:.4f},\n Recall: {recall:.4f}')
```

##### 200 estimator

```python
xgb_model200 = XGBClassifier(
    objective='multi:softprob',
    eval_metric='mlogloss',
    use_label_encoder=False,
    n_estimators=200,
    learning_rate=0.1,
    class_weight = 'balanced'

)
xgb_model200.fit(X_train, y_train_enc)
```

```python
y_true = y_val
eval_model_numeric_classes(xgb_model200,le, y_true, X_val)
```

```python
cf = ConfusionMatrixDisplay.from_estimator(
    xgb_model200, X_val, y_val_enc,display_labels = le.classes_, xticks_rotation = 90)
```

##### 500 estimator
